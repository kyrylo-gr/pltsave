```python
import matplotlib.pyplot as plt
import numpy as np
import cProfile
```

```python
x = np.linspace(0, 10, 1000)
y = np.linspace(0, 10, 1000)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

# y = np.sin(x) + np.random.normal(0, 0.1, len(x))


def create_plot():
    fig, ax = plt.subplots()
    ax.imshow(Z, extent=[0, 10, 0, 10], origin="lower", cmap="viridis")
    return fig, ax

create_plot()
```

```python
create_plot()
```

```python
cProfile.run("create_plot()")
```

```python
x, y, a, b, c, q, p = sp.symbols("x y a b c q p")

t = a * x**2 / 2 + 0 * y**2 / 2 + b * x * y
q_eq = sp.diff(t, x)
p_eq = sp.diff(t, y)

sol = sp.solve([q - q_eq, p - p_eq], [x, y])
x_eq, y_eq = sol[x], sol[y]

(q * x_eq + p * y_eq - t.subs({x: x_eq, y: y_eq})).simplify()
```
