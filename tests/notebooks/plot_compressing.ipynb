```python
CLASSIC_NAMES = {
    "L": "line",
    "A": "axes",
    "Le": "legend",
    "F": "fig",
    "LC": "line_coll",
    "T": "transform",
    "XA": "axis",
    "YA": "axis",
    "An": "annotation",
    "AI": "axim",
}
FORBIDDEN_KEYS = {"__name__", "children"}
COLOR_KEYS = {"color", "facecolor", "edgecolor",
              "markeredgecolor", "markeredgecolor"}

DEFAULT_COLORS = [
    "#1f77b4",
    "#ff7f0e",
    "#2ca02c",
    "#d62728",
    "#9467bd",
    "#8c564b",
    "#e377c2",
    "#7f7f7f",
    "#bcbd22",
    "#17becf",
]


def get_default_color(color: str | int) -> str:
    if isinstance(color, int) or (isinstance(color, str) and color.isdigit()):
        return DEFAULT_COLORS[int(color) % len(DEFAULT_COLORS)]
    return color


def convert_value(value: str, key: str) -> str:
    if key in COLOR_KEYS:
        return get_default_color(value)
    return value


def filter_non_none_values(d: dict):
    return {
        k: convert_value(v, k) for k, v in d.items() if v is not None and k not in FORBIDDEN_KEYS
    }


def dict_to_kwargs(d: dict) -> str:
    elms = []
    for k, v in d.items():
        if isinstance(v, str):
            elms.append(f"{k}='{v}'")
        else:
            elms.append(f"{k}={v}")

    return ", ".join(elms)


def find_name(expected_name: str, existed_names: set, start_index: int = 0):
    if start_index == 0:
        if expected_name not in existed_names:
            existed_names.add(expected_name)
            return expected_name
        else:
            start_index += 1

    for i in range(start_index, 100):
        if f"{expected_name}_{i}" not in existed_names:
            name = f"{expected_name}_{i}"
            existed_names.add(name)
            return name
    raise ValueError("Too many names")


def construct_fig(obj: dict, existed_names: set):
    name = find_name("fig", existed_names)
    code = f"{name} = plt.figure()\n"
    for child in obj.get("children", []):
        code += construct_obj(child, existed_names, name)
    return code


def construct_axes(obj: dict, existed_names: set, parent_name: str):
    name = find_name("ax", existed_names, 1)
    params = dict_to_kwargs(filter_non_none_values(obj))
    code = f"{name} = {parent_name}.add_axes(plt.Axes({parent_name}, {params}))\n"
    for child in obj.get("children", []):
        code += construct_obj(child, existed_names, name)
    return code


def construct_line(obj: dict, existed_names: set, parent_name: str):
    name = find_name("line", existed_names)
    params = dict_to_kwargs(filter_non_none_values(obj))
    code = f"{name} = {parent_name}.add_line(plt.Line2D({params}))\n"
    return code


def construct_obj(obj: dict, existed_names: set, parent_name: str | None = None):
    __name__ = obj["__name__"]
    if __name__ == "F":
        return construct_fig(obj, existed_names)
    if __name__ == "A":
        if parent_name is None:
            raise ValueError("Axes must have figure as a parent")
        return construct_axes(obj, existed_names, parent_name)
    if __name__ == "L":
        if parent_name is None:
            raise ValueError("Line must have axes as a parent")
        return construct_line(obj, existed_names, parent_name)
    return f"# {obj}"
```

```python
from typing import Final, Literal, Protocol
import plotly.offline as pyo


class ArticstJS(Protocol):
    def to_code(self, idx: int) -> str: ...


class FigureJS(ArticstJS):
    children: list[ArticstJS] | None = None
    __name__: Final = "F"

    def __init__(self, children: list, **kwargs):
        self.children = [loadObj(child) for child in children] if children else None

    def to_code(self, idx: int = 1):
        if not self.children:
            return ""
        children_code = "\n".join(
            [child.to_code(idx) for idx, child in enumerate(self.children, start=1)]
        )
        layout_code = "\n".join(
            [child.get_layout_code(idx) for idx, child in enumerate(self.children, start=1)]
        )
        return f"""import plotly.offline as pyo

fig = {{"data": [], "layout": {{}}}}
{children_code}
{layout_code}
pyo.iplot(fig)"""


class AxesJS(ArticstJS):
    children: list[ArticstJS] | None = None
    xlim: tuple[float, float] | None = None
    ylim: tuple[float, float] | None = None
    xlabel: str | None = None
    ylabel: str | None = None
    title: str | None = None
    rect: tuple[float, float, float, float] | None = None

    __name__: Final = "A"

    def __init__(self, **kwargs):
        self.xlim = kwargs.get("xlim")
        self.ylim = kwargs.get("ylim")
        self.xlabel = kwargs.get("xlabel")
        self.ylabel = kwargs.get("ylabel")
        self.title = kwargs.get("title")
        self.rect = kwargs.get("rect")
        children = kwargs.get("children")
        self.children = [loadObj(child) for child in children] if children else None

    def to_code(self, idx):
        traces_code = (
            "\n".join([child.to_code(idx) for child in self.children]) if self.children else ""
        )
        return f"{traces_code}"

    def get_layout_code(self, idx):
        if self.rect:
            x0, y0, width, height = self.rect
            x_domain = [x0, x0 + width]
            y_domain = [y0, y0 + height]
        else:
            x_domain = [0.0, 1.0]
            y_domain = [0.0, 1.0]

        layout_args = {
            "range": self.xlim,
            "title": self.xlabel,
            "domain": x_domain,
            "anchor": f"y{idx}",
        }
        xaxis_args = {k: v for k, v in layout_args.items() if v is not None}
        xaxis_code = f"fig['layout']['xaxis{idx}'] = {{{', '.join([f'{k!r}: {v!r}' for k, v in xaxis_args.items()])}}}"

        layout_args = {
            "range": self.ylim,
            "title": self.ylabel,
            "domain": y_domain,
            "anchor": f"x{idx}",
        }
        yaxis_args = {k: v for k, v in layout_args.items() if v is not None}
        yaxis_code = f"fig['layout']['yaxis{idx}'] = {{{', '.join([f'{k!r}: {v!r}' for k, v in yaxis_args.items()])}}}"

        title_code = f"fig['layout']['title'] = '{self.title}'" if idx == 1 and self.title else ""

        return f"{xaxis_code}\n{yaxis_code}\n{title_code}"


class LineJS(ArticstJS):
    xdata: list | None
    ydata: list
    color: str | None = None
    alpha: float | None = None
    linestyle: str | None = None
    linewidth: float | None = None
    marker: str | None = None
    markersize: float | None = None
    markerfacecolor: str | None = None
    markeredgecolor: str | None = None
    mode: Literal["lines", "markers", "lines+markers"] = "lines"

    __name__: Final = "L"

    def __init__(self, **kwargs):
        self.xdata = kwargs.get("xdata")
        self.ydata = kwargs.get("ydata", [])
        self.color = kwargs.get("color")
        self.alpha = kwargs.get("alpha")
        self.linestyle = kwargs.get("linestyle")
        self.linewidth = kwargs.get("linewidth")
        self.marker = kwargs.get("marker")
        self.markersize = kwargs.get("markersize")
        self.markerfacecolor = kwargs.get("markerfacecolor")
        self.markeredgecolor = kwargs.get("markeredgecolor")
        if self.linestyle is None:
            self.mode = "markers"
        elif self.marker is None:
            self.mode = "lines"
        else:
            self.mode = "lines+markers"
        

    def to_code(self, idx):
        trace_args = {
            "x": self.xdata,
            "y": self.ydata,
            "type": "scatter",
            "mode": self.mode,
            "line": {"color": self.color, "dash": self.linestyle, "width": self.linewidth}, 
            "marker": {
                "symbol": self.marker,
                "size": self.markersize,
                "color": self.markerfacecolor,
                "line": {"color": self.markeredgecolor},
            },
            "opacity": self.alpha,
            "xaxis": f"x{idx}",
            "yaxis": f"y{idx}",
        }
        filtered_trace_args = {k: v for k, v in trace_args.items() if v is not None}
        if "line" in filtered_trace_args:
            filtered_trace_args["line"] = {
                k: v for k, v in filtered_trace_args["line"].items() if v is not None
            }
        if "marker" in filtered_trace_args:
            filtered_trace_args["marker"] = {
                k: v for k, v in filtered_trace_args["marker"].items() if v is not None
            }
            if "line" in filtered_trace_args["marker"]:
                filtered_trace_args["marker"]["line"] = {
                    k: v for k, v in filtered_trace_args["marker"]["line"].items() if v is not None
                }

        trace_args_str = ", ".join([f"'{k}': {v!r}" for k, v in filtered_trace_args.items()])
        return f"fig['data'].append({{{trace_args_str}}})"


def loadObj(data: dict) -> ArticstJS:
    if not isinstance(data, dict):
        return data
    name = data["__name__"]
    if name == "F":
        return FigureJS(**data)
    if name == "A":
        return AxesJS(**data)
    if name == "L":
        return LineJS(**data)
    raise ValueError(f"Unknown object {name}")


# Example usage
line1 = LineJS(xdata=[1, 2, 3], ydata=[4, 5, 6], color="blue", linestyle="dash")
line2 = LineJS(xdata=[1, 2, 3], ydata=[6, 5, 4], color="red", linestyle="dot")
axes1 = AxesJS(
    children=[line1],
    xlabel="X Axis 1",
    ylabel="Y Axis 1",
    title="Title 1",
    rect=(0.0, 0.55, 1.0, 0.45),
)
axes2 = AxesJS(
    children=[line2],
    xlabel="X Axis 2",
    ylabel="Y Axis 2",
    title="Title 2",
    rect=(0.0, 0.0, 1.0, 0.45),
)
fig = FigureJS(children=[axes1, axes2])

print(fig.to_code())
```

```python
@np.vectorize
def func(x, y):
    return (
        1 * np.exp(-np.abs(x - y) ** 3 / 0.1)
        + np.exp(-np.abs(2 * x - y) ** 3 / 0.1)
        + np.exp(-np.abs(x - 2 * y + 10) ** 3 / 0.1)
        + np.exp(-np.abs(x / 2 - 2 * y + 5) ** 3 / 0.1)
        + np.random.normal(0, 2)
    )


x = np.linspace(0, 10, 1000)
y = np.linspace(0, 10, 3000)
X, Y = np.meshgrid(x, y)
Z = func(X, Y)
fig, ax = plt.subplots(1, 1)
im = ax.imshow(Z, origin="lower", extent=(0, 10, 0, 10))
cb = plt.colorbar(im)
```

```python
import scipy.ndimage
import numpy as np
from scipy.fftpack import dct, idct


def dct2(a):
    """Perform a 2D DCT (Discrete Cosine Transform)"""
    return dct(dct(a.T, norm="ortho").T, norm="ortho")


def idct2(a):
    """Perform a 2D inverse DCT (Inverse Discrete Cosine Transform)"""
    return idct(idct(a.T, norm="ortho").T, norm="ortho")


def compress_dct(image, keep_fraction=None, final_size=None):
    """
    Compress an image using the DCT by keeping only a certain fraction of the DCT coefficients.

    Args:
    image (numpy.ndarray): The 2D numpy array to compress.
    keep_fraction (float): The fraction of the DCT coefficients to keep.

    Returns:
    numpy.ndarray: The compressed image array.
    """
    # Perform the DCT
    dct_array = dct2(image)
    if keep_fraction is not None:
        expected_size = int((1 - keep_fraction) * dct_array.size)
    elif final_size is not None:
        expected_size = int(dct_array.size - final_size)
    else:
        raise ValueError("Either keep_fraction or final_size must be provided")
    # Determine how many coefficients to keep
    thresh = np.sort(np.abs(dct_array).ravel())[expected_size]

    # Zero out small coefficients and keep larger ones
    dct_array[np.abs(dct_array) < thresh] = 0

    return dct_array


def decompress_dct(compressed_dct):
    """
    Decompress an image that was compressed using the DCT.

    Args:
    compressed_dct (numpy.ndarray): The compressed DCT coefficients.

    Returns:
    numpy.ndarray: The decompressed image array.
    """
    # Perform the inverse DCT
    return idct2(compressed_dct)


# # Example of usage:
# # Create a sample 2D array (e.g., a grayscale image)
# image = np.random.rand(256, 256)  # Simulating a 256x256 pixel image

# # Compress the image
# # Keep only 5% of the DCT coefficients
# compressed_dct = compress_dct(image, keep_fraction=0.05)

# # Decompress the image
# reconstructed_image = decompress_dct(compressed_dct)

# print("Original Image Shape:", image.shape)
# print("Compressed Image Coefficients Shape:", compressed_dct.shape)
# print("Reconstructed Image Shape:", reconstructed_image.shape)


def apply_gaussian_filter(array, sigma, mode="reflect"):
    """
    Apply a Gaussian filter to a 2D numpy array.

    Args:
    array (numpy.ndarray): The 2D numpy array to filter.
    sigma (float): The standard deviation of the Gaussian kernel, controls the amount of blurring.

    Returns:
    numpy.ndarray: The blurred (filtered) 2D array.
    """
    return scipy.ndimage.gaussian_filter(array, sigma=sigma, mode=mode)
```

```python
def compress_dct_and_indices(image):

    # Store non-zero elements and their indices
    image = image.flatten()
    non_zero_indices = np.nonzero(image)[0]
    non_zero_values = image[non_zero_indices]
    return np.array([non_zero_indices, non_zero_values]).T.tolist()
```

```python
import unicodedata
import numpy as np
import aplot as ap
import sys
```

```python
encoding_map = {}
decoding_map = {}
number = 0
forbidden_symbols = (
    [" ", "?", "=", '"', "'", "&", ";", ",", ".", "_", "#", "\\"]
    + ["(", ")", "{", "}", "[", "]"]
    + [str(i) for i in range(10)]
    + [chr(i) for i in range(97, 123)]  # a-z
)
forbidden_categories = {"Mn"}

forbidden_code = set(ord(symbol) for symbol in forbidden_symbols)
lst = []
for i in range(0, 100_000):
    if i in forbidden_code:
        continue
    symbol = chr(i)
    if symbol.isprintable() and unicodedata.category(symbol) not in forbidden_categories:
        lst.append(i)
lst = np.array(lst)
```

```python
ap.ax().plot(np.diff(lst), label="diff").set(yscale="log")
```
